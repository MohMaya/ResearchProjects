### Dummy Threshold Calculator

```python
#print("Enter the Number of Pixels")
#n=int(input())

print("Enter the Value for Pixels")
px=list(map(int,input().split()))

for i in px:
    print(i,end='  ')
```

```python
from math import ceil
from math import pow
from math import sqrt
```

```python
card_px = len(px)
s=0
block_no = 1
for i in range(card_px-1):
   s = s + pow(abs(px[i]-px[i+1]),block_no)
#print(s)
s=s/(card_px-1)
#print(s)
s=ceil(s)
print(s)
```

### Making of graph

```python
grf = {}

for i in range(card_px):
    grf[i+1]=[]

for i in range(card_px):
    for j in range(card_px):
        if(i==j):
            continue
        if(abs(px[i]-px[j])<s):
            grf[i+1].append(j+1)
            
```

```python
for i in range(card_px):
    print("Nodes Related to node",i+1,":",grf[i+1])
```

```python
print(grf)
```

## Creating Blocks

```python

x_px=[]
y_px=[]
num_of_blocks=16

addend=I_dash.shape[0]//int(sqrt(num_of_blocks))
for i in range(0,I_dash.shape[0],addend):
    x_px.append(i)
    y_px.append(i)
x_px.append(I_dash.shape[0])
y_px.append(I_dash.shape[1])

print(x_px,y_px)
```

```python
blocks=[]
for j in range(len(x_px)-1):
        for k in range(len(y_px)-1):
            temp_block=I_dash[x_px[j]:x_px[j+1],y_px[k]:y_px[k+1]]
            blocks.append(temp_block)
```

```python
print(len(blocks))
```
