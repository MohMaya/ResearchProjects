```python
image = cv2.imread('gray1.jpg',cv2.IMREAD_GRAYSCALE)
image_dash = cv2.imread('gray1.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
a=plt.subplot(121)
a.imshow(image,cmap='gray')
a.set_title("Original")
plt.xticks([]), plt.yticks([])
b=plt.subplot(122)
b.imshow(image_dash, cmap = 'gray')
b.set_title("Modified")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
x_step_pixels=[]
y_step_pixels=[]
num_of_blocks=16
#Assuming the image to be square shaped
step_size=image_dash.shape[0]//int(sqrt(num_of_blocks))
for i in range(0,image_dash.shape[0],step_size):
    x_step_pixels.append(i)
    y_step_pixels.append(i)
x_step_pixels.append(image_dash.shape[0])
y_step_pixels.append(image_dash.shape[1])

print(x_step_pixels,y_step_pixels)
```

```python
print(image_dash.shape)
image_doubDash = image_dash.flatten()
print(image_doubDash.shape)
blocks=[]
num_of_blocks=16
for i in range(num_of_blocks):
    f=image_doubDash.shape[0]//num_of_blocks
    print(f*i,"to",f*(i+1))
    temp_block=image_doubDash[f*i:f*(i+1)]
    blocks.append(temp_block)
```

```python
print(len(blocks))
```

```python
def s_calculator(pixels_in_the_block,block_num):
    card_px = len(pixels_in_the_block)
    s=0
    for i in range(card_px-1):
       s = s + abs(pow(pixels_in_the_block[i],block_num)-pow(pixels_in_the_block[i+1],block_num))
    #print(s)
    s=s/(card_px-1)
    s=pow(s,1/block_num)
    #print(s)
    s=ceil(s)
    return s
```

```python
s_values = []
for j in range(len(blocks)):
    s_values.append(s_calculator(blocks[j],j+1))
print(s_values)
```

```python
def graph_maker(block,s):
    graph = {}
    cardinality_block = len(block)
    
    for i in range(cardinality_block):
        graph[i] = []
    for i in range(cardinality_block):
        for j in range(cardinality_block):
            if(i==j):
                continue
            elif(abs(int(block[i])-int(block[j]))<s):
                graph[i].append(j)
        if(((i*100)/cardinality_block)%10 == 0):
            print('=',end="")
    print('')
    return graph
                
```

```python
graph_of_first_block = graph_maker(blocks[0],s_values[0])
```
