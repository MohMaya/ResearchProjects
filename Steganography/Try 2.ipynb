```python
def s_calculator(pixels_in_the_block,block_num):
    card_px = len(pixels_in_the_block)
    s=0
    for i in range(card_px-1):
       s = s + abs(pow(pixels_in_the_block[i],block_num)-pow(pixels_in_the_block[i+1],block_num))
    #print(s)
    s=s/(card_px-1)
    s=pow(s,1/block_num)
    #print(s)
    s=ceil(s)
    return s
```

```python
s_values = []
for j in range(len(blocks)-1):
    s_values.append(s_calculator(blocks[j],j+1))
print(s_values)
```
